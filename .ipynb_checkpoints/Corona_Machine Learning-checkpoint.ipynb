```python
cat_col =['Province/State','Country']
num_col = ['day', 'month']
```

```python
gr = GradientBoostingRegressor(n_estimators =600)
```

```python
X = virus_corona[cat_col + num_col]
y =  virus_corona['Confirmed'] 

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

categorical_columns = ['Province/State','Country','day', 'month']
numerical_columns = []
categorical_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])


preprocessing = ColumnTransformer(
    [('cat', categorical_pipe, categorical_columns)])

rf = Pipeline([
    ('preprocess', preprocessing),
    ('feature_selection', SelectFromModel(gr, prefit=False)),
    ('classifier', gr)
])
```

```python
categorical_columns = ['Province/State','Country','Last Update']
numerical_columns = ['Confirmed','Deaths','Recovered']
categorical_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
numerical_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='median'))
])
preprocessing = ColumnTransformer(
    [('cat', categorical_pipe, categorical_columns),
     ('num', numerical_pipe, numerical_columns)],
)
rf = Pipeline([
    ('preprocess', preprocessing),
 
    ('classifier',  KMeans(n_clusters=4))
])
```

```python
rf.fit(virus_corona)
```
